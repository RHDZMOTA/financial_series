# Example for type currency

```python
%matplotlib inline

from forex_data import *

import matplotlib.pyplot as plt
import datetime as dt
import pandas as pd
import numpy as np
```

```python
# Initialize currency
a = currency(units='EUR', base = 'USD', t0 = '2015/01/01', tf = '2016/01/01')

# Download prices and calculate returns
a.fill()
```

```python
# Show plots for price and returns; 
a.prices.plot(); a.returns.plot()
```

```python
# Head of prices
a.prices.head(10)
```

```python
# Tail of prices
a.prices.tail(10)
```

```python
# Description 

print('Prices: \n\n',a.prices.Adj_close.describe())

print('\n\nReturns: \n\n',a.returns.Adj_close.describe())
```

```python
# Difference among the open price and the close price. 

plt.figure()
(a.prices.Open - a.prices.Close).plot()
plt.title('Difference between open and close price')
plt.ylabel('Value in euros')
plt.show()
```
